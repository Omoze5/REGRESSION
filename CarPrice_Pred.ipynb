# import the necessay libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("CarPrice_Assignment.csv")
data.head(5)
```

```python
numeric_columns = data[["symboling","wheelbase", "carlength", "carwidth", "carheight","curbweight","enginesize", "boreratio","stroke", "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg" ]]
categprical_columns = data[[ "CarName", "fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "enginelocation", "enginetype", "cylindernumber"]]
```

```python
le = StandardScaler()
le.fit(numeric_columns)
scaled_numeric_columns = le.transform(numeric_columns)
```

```python
encoded_categorical_columns = pd.get_dummies(categprical_columns)
```

```python
encoded_categorical_columns
```

```python
x = pd.concat([encoded_categorical_columns,pd.DataFrame(scaled_numeric_columns)], axis = 1)
x.columns = x.columns.astype(str)
```

```python
x
```

## Dimensionality reduction was carried out to reduce dimensions and also enhance efficiency of the model

```python
y = data["price"]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
x_train, x_test,y_train, y_test = train_test_split(x_pca, y, test_size= 0.2 )
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy = r2_score(y_test, y_pred)
accuracy
```

```python
regression_line = np.polyfit(y_pred, y_test, 1)
plt.plot(y_test, np.polyval(regression_line, y_test), color = "green", label= "Regression Line")
plt.xlabel("The Y variable")
plt.ylabel("The Predicted Value")
plt.legend()
```
